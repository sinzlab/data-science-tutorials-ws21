## Loops (enumerate and zip)

Very often in programming we are faced with a situration where we need to perform iterative processes. This can either include 
- going through all the elements of a collection (e.g. list, tuple, etc.), also called an iterable object, and perform some operation on the elements -> we saw one situation before where we wanted to increase every element of our list by 2
- or doing a specific operation repeatedly until a certain outcome is achieved. 

Iterative processes in Python (or any programming language) are called **loops**.

Generally speaking, there are two types of loops: **for loops** and **while loops**.

- **for loops** are the ones that you say "I care about the number of iterations, so do something for a specific number of times"
- when using **while loops** on the other hand, we may not necessarily care about the number of iteration, but about a specific condition. So we basically say "do the iteration until something happened."

Let's talk about for loops first.

In the case of **for loops** generally two situations can happen: One situation is when you have a range and you want to swip through this range. Let's say I now I want to start from 1, and I want to end before 10 (that is 9). So in this case, we need some way to express our range of numbers.. we can do that in Python via `range()`. So if we say `range(1, 10)` we mean we want to go from 1 up to 10, but not 10 - 10 is not included!

And ultimately, for each number in this range, we want to do something with the number (e.g., print it).

How do we do this in Python?

``` python
for num in range(1, 10)
# Do something
```

So this "Do something" is essentially a block of code that should only happen inside this loop - it belongs to the loop. In Python we should also express this *belonging* somehow. and that is done by adding a **colon** and **indentation**.

So our code above will change to:
``` python
for num in range(1, 10):
    # Do something
```

And we are done! Let's replace "Do something" with `print(num)` and see what happens:

````python
for num in range(1, 10):
    print(num)
````

And now we can increase every element of our list by 2, right?

Let's do it:

````python
a_list = [1, 2, 3, 4, 5]
for ind in range(0, len(a_list)):
    a_list[ind] = a_list[ind] + 2
````

````python
a_list
````

**Fun fact**: Same as indexing, we dont really have to say that we want to start from 0, because starting points are 0 by default in Python. So `range(0, 10)` is same as `range(10)`.

We did not talk about the other way with which we can do for loops. And that is to directly iterate through the elements of a list, instead of indexing, which in many case is in fact the preferable way (the pythonic way). Most of the time what we really want to do, is to do something with every element of the list.. In Python, we can in fact skip any indexing and directly iterate through a list (or any collection):

````python
a_list = [5, 2, 3, 4, 5]
for val in a_list:
    print(val)
````

````python
a_list
````

### `enumerate()`

Note that in this case we do not have any index whatsoever, but we have direct access to the elements. What if we wanted to directly access the element of one list, but we also needed the index to access the elements of another list, for instance? - for this we can use `enumerate()`.

````python
a_list = [1, 2, 3, 4, 5]
b_list = [5, 4, 3, 2, 1]

for ind, val in enumerate(a_list):
    print(val, b_list[ind])
````

Before we move on to some exercises, let's also shortly talk about while loop.

````python
counter = 0
while counter < 5:
    print(counter)
    counter += 1
````

#### Some Important Points

* Note that by using if, elif, and else we are imposing a priority on conditions:

````python
a = 5
b = 3

if a > b:
    print("a is bigger than b")
elif a > 3:
    print("a is bigger than 3")
else:
    print("a is smaller than both 3 and b")
````

Note that both of the following conditions are True:
- a > b
- a > 3

But, since `if` has priority over `elif` (or any subsequent condition), it runs the code under `if`.

* Also note that we can get rid of this priority by using multiple `if` statements

````python
a = 5
b = 3

if a > b:
    print("a is bigger than b")
    
if a > 3:
    print("a is bigger than 3")
else:
    print("a is smaller than both 3 and b")
````

When we use an `if`, then the following `elif` and `else` statements are attached to it, and prioritized accordingly.

**Note**: I am using capitalized "and", "or", and "not" keywords to make a distinction between a Python keyword and a normal word. In python these are all lowercase.

- The expression x AND y is True only if x and y both are True.
- The expression x OR y is False only if x and y both are False.

```python
something_false and x -> something_false
something_true or x -> something_true
something_false or x -> x
```

While using the boolean operations, the following values are interpreted as false: False, None, numeric zero of all types, and empty strings and collections (e.g. tuples, lists). All other values are interpreted as true.

## Functions

It happens often that you have a specific set of instructions and you want to perform the same thing at different places in your code. Of course, you could always write the whole thing over and over, but another approach is to pack that set of instructions, give it a name, and just use that name to perform it, instead of re-writing everything everytime - **keep your code dry!** And **functions** help you do exactly that.

Functions are a set of instructions (lines of code) packaged together that perform a specific task, and return the result(s). Usually a function also get inputs (referred to as **function arguments**), and in Python you can define functions with any number of arguments, including no arguments.

To create a function we use the Python keyword `def`:

```python
def my_func(arg):
    # do something
    # result = final thing
    return result
```

````python
def my_func():
    result = 5
    return result
````

````python
my_func()
````

````python
def add_3(input_val):
    result = input_val + 3
    return result
````

````python
add_3(5)
````

Note that the variables defined in the function are not available outside it

````python
result
````

````python
def add_two_values(a, b):
    result = a + b
    return result
````

````python
add_two_values(5, 3)
````

````python
add_two_values(5)
````

````python
def add_two_values(a, b=3):
    result = a + b
    return result
````

````python
add_two_values(5)
````

Two types of arguments:
- **Positional**: have no default arguments and must be passed to the function 
- **Optional**: you guessed it - have a default arguments which will be used in case no value is passed by the user

Functions can also return **multiple outputs**

````python
def add_and_substract_two_values(a, b=3):
    result_add = a + b
    result_sub = a - b
    return result_add, result_sub
````

````python
add_and_substract_two_values(5)
````

````python
res1, res2 = add_and_substract_two_values(5)
````

````python
res1
````

### Anatomy of matplotlib plot

In essence, there are two big components that you need to take into account:

- The **Figure** is the overall window or page that everything is drawn on. It’s the top-level component of all the ones that you will consider in the following points. You can create multiple independent Figures. A Figure can have several other things in it, such as a **suptitle**, which is a centered title to the figure. You’ll also find that you can add a **legend** and **color bar**, for example, to your Figure.

- To the figure you add Axes. The Axes is the area on which the data is plotted with functions such as `plot()` and `scatter()` and that can have ticks, labels, etc. associated with it. This explains why Figures can contain multiple Axes.
<div style="text-align:center">
    <a href="https://matplotlib.org/1.5.1/faq/usage_faq.html">
    <img src ="https://matplotlib.org/1.5.1/_images/fig_map.png" />
</div>

Pretty much whatever we did above can be done with the axis objec as well. In other words, the methods used to manipulate the properties of a plot, for example `plt.xlabel()` all exist for an axes object as well. For instacne, instead of `plt.xlabel()` we can also use `ax.set_xlabel()`.  

````python
fig, ax = plt.subplots()  # figsize=(10, 5)

# remove axis boarders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(x, y1, label='sin', color='r', linestyle='-.', linewidth=5)
ax.plot(x, y2, label='cos', color='g', linestyle='-.', linewidth=5)

# setting labels
ax.set_xlabel('', fontsize=30)  # plt.xlabel('X Label')
ax.set_ylabel('Y Label', fontsize=30);  # plt.ylabel('Y Label')

# changing the values on the axes
ax.locator_params(axis='both', nbins=4)
ax.set_xticks([90, 180, 300])
ax.set_xticklabels(['Python', 'Is', 'Awesome'], rotation=30)
ax.tick_params(axis='both', 
               direction='in', 
               labelsize=30)

ax.legend(loc='lower left')
ax.grid()
````

````python
x = np.arange(0, 360, 5)
x_rad = np.deg2rad(x)
y1 = np.sin(x_rad)
y2 = np.cos(x_rad)

fig, axes = plt.subplots(nrows=3, ncols=3, sharex=True, sharey=True, figsize=(15, 10))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'b']
for ind, (col, ax) in enumerate(zip(colors, axes.flat)):
    ax.plot(x, y1 + np.random.random(x.shape[0]), color=col, linewidth=ind+1, label= 'Line width = ' + str(ind))

    # changing the values on the axes
    ax.locator_params(axis='both', nbins=4)
    ax.tick_params(axis='both', labelsize=30)
    ax.legend(loc='lower left', fontsize=15)
    ax.grid()

# Note that we are setting the labels for specific axes
axes[2, 1].set_xlabel('X Label', fontsize=30)
axes[1, 0].set_ylabel('Y Label', fontsize=30)
fig.tight_layout()
````

````python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(15, 10))

gs = GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, :-1])
ax3 = plt.subplot(gs[1:, -1])
ax4 = plt.subplot(gs[2, 0])
ax5 = plt.subplot(gs[2, 1])

axes = (ax1, ax2, ax3, ax4, ax5)

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'b']
for ind, (col, ax) in enumerate(zip(colors, axes)):
    ax.plot(x, y1 + np.random.random(x.shape[0]), color=col, linewidth=ind+1, label= 'Line width = ' + str(ind))

    # changing the values on the axes
    ax.locator_params(axis='both', nbins=4)
    ax.tick_params(axis='both', labelsize=30)
    ax.legend(loc='lower left', fontsize=15)
    ax.grid()

fig.tight_layout()
````
